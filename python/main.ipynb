json形式では扱いずらいので、csv形式に変換を行う

````python
def json_2_csv():
    # /dataからjsonファイルを読み込み
    json_files = [f for f in os.listdir('data') if f.endswith('.json')]
    for json_file in json_files:
        csv_file = 'data/' + json_file.replace('.json', '.csv')
        if not os.path.exists(csv_file):
            json_data = json.load(open('data/' + json_file, 'r', encoding='utf-8'))
            csv_data = json_parse(json_data)
            csv_data.to_csv(csv_file, index=False)

"""
```json
[
  {
    "blendShapes" : {
      "browDown_L" : 0,
      "browDown_R" : 0,
      "browInnerUp" : 0.3971520960330963,
      "browOuterUp_L" : 0.15014605224132538,
      "browOuterUp_R" : 0.15536083281040192,
      "cheekPuff" : 0.02889101952314377,
      "cheekSquint_L" : 0.06382670998573303,
      "cheekSquint_R" : 0.06006058678030968,
      "eyeBlink_L" : 0.0076168617233633995,
      "eyeBlink_R" : 0.007428889628499746,
      "eyeLookDown_L" : 0,
      "eyeLookDown_R" : 0,
      "eyeLookIn_L" : 0.5230473279953003,
      "eyeLookIn_R" : 0,
      "eyeLookOut_L" : 0,
      "eyeLookOut_R" : 0.4302368462085724,
      "eyeLookUp_L" : 0.377517431974411,
      "eyeLookUp_R" : 0.38386932015419006,
      "eyeSquint_L" : 0.15236200392246246,
      "eyeSquint_R" : 0.15236683189868927,
      "eyeWide_L" : 0.0026134755462408066,
      "eyeWide_R" : 0.0029545845463871956,
      "jawForward" : 0.07849083095788956,
      "jawLeft" : 0,
      "jawOpen" : 0.14967109262943268,
      "jawRight" : 0.007712039165198803,
      "mouthClose" : 0.1480281502008438,
      "mouthDimple_L" : 0.05187155306339264,
      "mouthDimple_R" : 0.057662129402160645,
      "mouthFrown_L" : 0.046667203307151794,
      "mouthFrown_R" : 0.06657587736845016,
      "mouthFunnel" : 0.07630263268947601,
      "mouthLeft" : 0,
      "mouthLowerDown_L" : 0.1017029881477356,
      "mouthLowerDown_R" : 0.10114526003599167,
      "mouthPress_L" : 0.053885020315647125,
      "mouthPress_R" : 0.04836598038673401,
      "mouthPucker" : 0.10161589831113815,
      "mouthRight" : 0.0004691672511398792,
      "mouthRollLower" : 0.08914703130722046,
      "mouthRollUpper" : 0.044839341193437576,
      "mouthShrugLower" : 0.09597064554691315,
      "mouthShrugUpper" : 0.04718717560172081,
      "mouthSmile_L" : 0,
      "mouthSmile_R" : 0,
      "mouthStretch_L" : 0.22257839143276215,
      "mouthStretch_R" : 0.23614069819450378,
      "mouthUpperUp_L" : 0.027168704196810722,
      "mouthUpperUp_R" : 0.027438458055257797,
      "noseSneer_L" : 0.0971347838640213,
      "noseSneer_R" : 0.12149970978498459,
      "tongueOut" : 1.3884564609156769e-08
    },
    "timestamp" : 769780917.311326
  }
]
```
"""

def json_parse(json_data):
    # 各要素のblendShapesを展開し、timestampと結合してDataFrame化
    records = []
    for entry in json_data:
        row = entry["blendShapes"].copy()
        row["timestamp"] = entry["timestamp"]
        records.append(row)
    df = pd.DataFrame(records)
    return df

json_2_csv()
````

# 可視化関数

````python
# 時系列データをグラフとして描画
def plot_blend_shapes(df):
    fig = plt.figure(figsize=(40, 12))
    ax = fig.add_subplot(1, 1, 1)
    df.set_index('timestamp').plot(ax=ax)
    # グラフのタイトルとラベルを設定
    ax.set_title('BlendShapes Time Series')
    ax.set_xlabel('Timestamp')
    ax.set_ylabel('BlendShapes Values')
    # x軸の目盛りを調整
    ax.xaxis.set_major_locator(plt.MaxNLocator(10))  # 最大10個の目盛り
    # 軸のメモリは分・秒単位で表示
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x, unit='s').strftime('%H:%M:%S')))
    plt.xticks(rotation=45)  # x軸の目盛りを45度回転
    # グラフを表示
    plt.tight_layout()  # レイアウトを調整
    plt.show()
````

````python
# 顎の動きのグラフを描画
def plot_jaw_movement(df):
    fig, ax = plt.subplots(figsize=(40, 6))
    df.set_index('timestamp')[['jawOpen']].plot(ax=ax)
    # グラフのタイトルとラベルを設定
    ax.set_title('Jaw Movement Over Time')
    ax.set_xlabel('Timestamp')
    ax.set_ylabel('Jaw Movement Values')
    # x軸の目盛りを調整
    ax.xaxis.set_major_locator(plt.MaxNLocator(10))  # 最大10個の目盛り
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x, unit='s').strftime('%H:%M:%S')))
    plt.xticks(rotation=45)  # x軸の目盛りを45度回転
    # グラフを表示
    plt.tight_layout()  # レイアウトを調整
    plt.show()
````

# データの読み込み

````python
for csv_file in os.listdir('data'):
    if csv_file.endswith('.csv'):
        df = pd.read_csv('data/' + csv_file)
        df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
        plot_blend_shapes(df)
        plot_jaw_movement(df)
````
